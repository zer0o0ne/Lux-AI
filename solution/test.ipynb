```python
def update(node, v):
    node.N += 1
    node.V += v
    node.Q, node.u = node.V / node.N, node.T * node.p / (1 + node.N)
    if node.parent is not None:
        update(node.parent, v)

class Node:
            def __init__(self, p, parent, state, T = 1):
                self.p, self.u, self.T = p, T * p, T
                self.parent, self.state = parent, state
                self.Q, self.N, self.V = 0, 0, 0
                self.children = []

            def make_root(self):
                self.parent = None

            def get_probabilities(self):
                numbers = [child.N for child in self.children] 
                sum = np.sum(numbers)
                return [n / sum for n in numbers]

            def get_leaf(self):
                if len(self.children) == 0: return self
                scores = [child.Q + child.u for child in self.children]
                return self.children[np.argmax(scores)].get_leaf()

            def inspect(self, P, states, v):
                assert len(P) == len(states)
                for i in range(len(P)):
                    self.children.append(Node(P[i], self, states[i], self.T))
                update(self, v)
```
